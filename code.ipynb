**Importing all dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  


```

**Importing Dataset**

```python
data=pd.read_csv('Python Airbnb Project Dataset.csv', encoding_errors='ignore')
```

**Data Cleaning**

```python
#finding null values
data.isnull().sum()

#removing null values
data.dropna(inplace=True)

#after removing null values
data.isnull().sum()
```

```python
#finding duplicate values
data.duplicated().sum()
data[data.duplicated()]
```

```python
#removing duplicates
data.drop_duplicates(inplace=True)

#after removing duplicates
data.duplicated().sum()
```

```python
#verifying data types
data.dtypes

#changing some data types
data['id'] = data['id'].astype(object)
data.dtypes

data['host_id']=data['host_id'].astype(object)
data.dtypes
```

**EDA**


**Data Analysis**

**Univariate Analysis**

```python
#identifying outliers in prize
df=data[data['price']<3000]

sns.boxplot(data=df, x='price')
plt.show()
```

```python
#price distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.ylabel("Frequency")
plt.title('Frequency of price')
plt.show()
```

```python
#
plt.figure(figsize=(6,3))
sns.histplot(data=df, x='availability_365', bins=30)
plt.ylabel("Frequency")
plt.title('Availability of rooms')
plt.show()
```

```python
#comparing neighbourhood_group with average price
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**


```python
#creating a new column price_per_bed
df['price_per_bed']=df['price']/df['beds']
df.head()
```

```python
#average price per bed
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

**Bivariate Analysis**

```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.title('Price of room type')
plt.show()
```

```python
#no.of.reviews with price relationship
plt.figure(figsize=(8,5))
plt.title('No.of.reviews with price')
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```

```python
#relationship between multiple variables
data.dtypes
sns.pairplot(data=df,vars=('price','minimum_nights','availability_365'),hue='room_type')
plt.show()
```

```python
#geographical distribution
plt.figure(figsize=(10,4))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical Listings')
plt.show()
```

```python
#heatmap--correlation between multiple variables
corr=df[['latitude','longitude','price','minimum_nights','number_of_reviews','availability_365','beds']].corr()
corr
sns.heatmap(data=corr,annot=True)
```
